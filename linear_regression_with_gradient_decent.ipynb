```python
import numpy as np

# Sample data: house sizes (X) and prices (y)
X = np.array([1, 2, 3])  # House sizes (e.g., in 1000 sqft)
y = np.array([2, 4, 6])  # Prices (e.g., in $100,000)

# Initialize slope and intercept (start with 0)
slope = 0
intercept = 0

# Learning rate (step size)
learning_rate = 0.01

# Number of steps to adjust the line
num_steps = 10000

# Keep track of errors (optional)
errors = []

for _ in range(num_steps):
    # Predict prices using current slope/intercept
    predictions = slope * X + intercept
    
    # Calculate error (difference from actual prices)
    error = y - predictions
    
    # Update slope and intercept using gradient descent
    # (This is where the "magic" happens)
    slope += learning_rate * (1/len(X)) * np.sum(error * X)
    intercept += learning_rate * (1/len(X)) * np.sum(error)
    
    # Track error (optional)
    current_error = np.mean(error ** 2)
    errors.append(current_error)
    print(f"Best slope: {slope:.2f}, Best intercept: {intercept:.2f}")
print(f"Best slope: {slope:.2f}, Best intercept: {intercept:.2f}")
```

```python
# Data: x and y values
x = [1, 2, 3]
y = [2, 4, 6]

# Start guessing
slope = 0
intercept = 0
learning_rate = 0.1

for _ in range(100):  # Try 100 times
    # Predict y for all x
    predictions = [slope * xi + intercept for xi in x]
    
    # Calculate errors
    errors = [yi - pred for yi, pred in zip(y, predictions)]
    
    # Adjust slope and intercept
    slope += learning_rate * sum(err * xi for err, xi in zip(errors, x)) / len(x)
    intercept += learning_rate * sum(errors) / len(x)

print(f"Final slope: {slope}, Final intercept: {intercept}")
```
